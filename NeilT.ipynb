```python
#Load Dependencies
import pandas as pd
```

```python
#Extract csv
file = './Resources/alcoholSubstanceAbuse.csv'
df = pd.read_csv(file)
```

```python
#Change column names
df = df.rename(columns={'Location':'Country_ID','Period':'Year','Indicator':'Indicator',
                'Dim1': 'Sex', 'First Tooltip': 'First Tooltip'})
df.head()
```

```python
#Group by year of 2018
new_df = df.loc[df['Year']==2018]
```

```python
#Export cleaned dataframe to csv in Resources folder
```

```python
new_filepath = './Resources/alcoholSubstanceAbuse_clean.csv'
new_df.to_csv(new_filepath)
```
