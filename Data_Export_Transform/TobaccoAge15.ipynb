```python
import pandas as pd
```

```python
csv_path = "resources/tobaccoAge15.csv"

tobacco_df = pd.read_csv(csv_path, encoding="utf-8")
tobacco_df.head()
```

```python
renamed_df = tobacco_df.rename(columns={"Location": "Country_Id",
                                        "Period": "Year",
                                       "Dim1": "Sex"})
renamed_df.head()
```

```python
clean_df = renamed_df[renamed_df['Year'] == 2018]
clean_df
```

```python
clean_df.to_csv("resources/tobaccoAge15_clean.csv",
                  encoding="utf-8", index=False, header=True)
```
